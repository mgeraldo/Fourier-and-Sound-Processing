```python
import pyaudio
import numpy as np
import struct
import matplotlib.pyplot as plt
#%matplotlib tk
```

```python
CHUNK = 1024 
#FORMAT = pyaudio.paInt16
FORMAT = pyaudio.paFloat32
CHANNELS = 1
RATE = 44100

p = pyaudio.PyAudio()

stream = p.open(
    format=FORMAT,
    channels=CHANNELS,
    rate=RATE,
    input=True,
    output=True,
    frames_per_buffer=CHUNK
)

data = stream.read(CHUNK)
```

```python
data
```

```python
#data_int_B = struct.unpack(str(2*CHUNK)+ 'B',data) #'B' is for 1 byte integers
#data_int_i = struct.unpack(str(2*CHUNK)+ 'i',data) #'B' is for 1 byte integers
data_int_f = struct.unpack(str(CHUNK)+ 'f',data) # 'f' is for 4 byte Floats 
```

```python
fig, ax = plt.subplots()
ax.plot(data_int_f[0:1000])
plt.show()
```
